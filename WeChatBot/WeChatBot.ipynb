```python
'''
Check income messages
'''
my_msgs = bot.messages
if my_msgs:
    print('You have '+ str(len(my_msgs) ) + '. Here is the first message: ' + str(my_msgs[0]) )
else:
    print('No messages since you have logged in with Python')

'''
Search for your messages
'''
my_search = bot.messages.search('Python', sender=bot.self)
print('Found ' + str(len(my_search)) + ' messages. \nMessages:\n ' + str(my_search) )
```

```python
import schedule
import time

def job():
    my_wc_group[0].send("Hello")

# schedule.every().day.at("10:30").do(job)
schedule.every(10).seconds.do(job) 

cc = 0
while 1:                                
    schedule.run_pending()
    time.sleep(1)
    cc += 1
    if cc==10:
        break
```
