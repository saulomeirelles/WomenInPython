# Part 1: Traditional Programming vs Machine Learning

```python
'''
Here we import the packages we need for finding x and y.
'''
import tensorflow as tf
from tensorflow import keras

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import os
from glob import glob
```

```python
'''
Traditional Programming
'''

x = np.array([1,2,3,4,5],float)
y = np.array([2,4,6,8,10],float)

print('y divided by x is equal to ', y/x)
```

```python
'''
Now the we now the rule, we can create a function to find ys if xs=10.
'''
def find_y(x):
    y = 2*x
    return y

print(find_y(10))

```

```python
'''
Machine Learning
'''
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
model.fit(x,y,epochs=500)
```

```python
print(model.predict([10.0]))
```

# Part 2: Simple image recognition

```python
# (training_images, training_labels), (test_images, test_labels) = mnist.load_data()
data_train_file = "fashion-mnist_train.csv"
data_test_file  = "fashion-mnist_test.csv"

df_train = pd.read_csv(data_train_file) 
df_test = pd.read_csv(data_test_file)     
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
training_images = df_train.drop("label", axis=1) / 255.0 # normalize between 0 and 1
training_labels = df_train["label"]

test_images = df_test.drop("label", axis=1) / 255.0 # normalize between 0 and 1
test_labels = df_test["label"]
```

```python
training_labels.value_counts().plot.bar()
```

```python
df_labels = pd.DataFrame(data=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot'], columns=['Description']).rename_axis("Label")
df_labels
```

```python
print(training_images.shape)
print(test_images.shape)
```

```python
dim = 28
idx = 1
plt.imshow(training_images.iloc[idx].values.reshape(dim,dim))
print(training_labels[idx])
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu), 
                                    tf.keras.layers.Dense(10,  activation=tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.train.AdamOptimizer(),
              loss      = 'sparse_categorical_crossentropy',
              metrics   = ['accuracy'] )

model.fit(training_images.values, training_labels.values, epochs=5)
```

```python
model.evaluate(test_images.values, test_labels.values)
```

```python
idx = 0
plt.imshow(test_images.iloc[idx].values.reshape(dim,dim))
print('label: ', test_labels[idx])
print( 'validation :', model.predict_classes( np.expand_dims( test_images.iloc[idx].values.reshape(dim,dim), axis=0) ) )
```

```python
'''
images retrieved from the Internet
'''
fig, ax = plt.subplots(1,3, figsize=(8,3), sharey=True)
for i in range( np.shape( glob('*.jpeg') )[0] ):
    im = mpimg.imread( glob('*.jpeg')[i] )
    ax[i].imshow(im, aspect='auto')
```

```python
 def prep_img(image_file_path):
    image_string = tf.read_file(image_file_path)
    image        = tf.image.decode_jpeg(image_string, channels=3)
    image        = 1 - tf.image.rgb_to_grayscale(image)
    image        = tf.image.convert_image_dtype(image, dtype=tf.float32)
    image        = tf.image.resize_image_with_pad(image, 28, 28)
    image.set_shape([28, 28, 1])
    return image
```

```python
fig, ax = plt.subplots(1,3, figsize=(8,3), sharey=True)

for i in range( np.shape( glob('*.jpeg') )[0] ):

    image = prep_img( glob('*.jpeg')[i] )
    init  = tf.global_variables_initializer()    
    sess  = tf.Session()
    sess.run(init)

    image_array = np.fliplr( image.eval(session=sess).squeeze() )
    
    pred = model.predict_classes( np.expand_dims( image_array, axis=0 ) )
    print('The predicted label is ' , pred)
    print('This label corresponds to: ', df_labels.Description[pred].values)
    
    ax[i].imshow( image_array )
    ax[i].set_title( ''.join(df_labels.Description[pred].values) )
```

```python
fig, axs = plt.subplots(1,3,figsize=(20, 6))

axs[0].imshow(  im, aspect="auto")
axs[1].imshow(  image.eval(session=sess).squeeze(), cmap=plt.cm.Blues_r, aspect="auto") 
axs[2].matshow( image.eval(session=sess).squeeze(), cmap=plt.cm.Blues_r, aspect="auto") 

for (j, i), v in np.ndenumerate( np.asarray( image.eval(session=sess).squeeze() ) ):
    axs[2].text( i, j, str( int(round(10*v)) ),
               horizontalalignment='center',
               verticalalignment='center')

tls = ['How we see','Processed image','How computers see']
for i,ax in enumerate(axs):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.grid()
    if i < 2:
        ax.set_title(tls[i], fontsize=40, color='black', pad=30)
    else:
        ax.set_title(tls[i], fontsize=40, color='black', pad=15)
        
fig.tight_layout()
# fig.savefig('../input/human-computer-vision.png')
```
